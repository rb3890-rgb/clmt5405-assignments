## Part III: Functions

### 1) Write a function to convert temperature from Kelvin to Celsius

…and another to convert from Celsius to Kelvin.

````python
def kelvin_to_celsius(t_k):
    """Convert a temperature from Kelvin to Celsius."""
    return t_k - 273.15

def celsius_to_kelvin(t_c):
    """Convert a temperature from Celsius to Kelvin."""
    return t_c + 273.15
````

### 2) Write a function to convert temperature to Fahrenheit

Include an optional keyword argument to specify whether the input is in Celsius or Kelvin. Call your previously defined functions if necessary.

````python
def to_fahrenheit(temp, scale="celsius"):
    """Convert a temperature to Fahrenheit.

    scale : "celsius" (default) or "kelvin" -- the scale of `temp`.
    """
    if scale == "kelvin":
        temp = kelvin_to_celsius(temp)     # reuse the function above
    elif scale != "celsius":
        raise ValueError("scale must be 'celsius' or 'kelvin'")
    return temp * 9 / 5 + 32
````

### 3) Check that the outputs are sensible

by trying a few examples

````python
print(to_fahrenheit(0))                        # 0 C   -> 32 F
print(to_fahrenheit(100))                      # 100 C -> 212 F
print(to_fahrenheit(273.15, scale="kelvin"))   # 0 C   -> 32 F
print(to_fahrenheit(373.15, scale="kelvin"))   # 100 C -> 212 F
````

### 4) Now write a function that converts _from_ Fahrenheit

Use a keyword argument to specify whether you want the output in Celsius or Kelvin.

````python
def from_fahrenheit(temp_f, scale="celsius"):
    """Convert a Fahrenheit temperature to Celsius (default) or Kelvin."""
    celsius = (temp_f - 32) * 5 / 9
    if scale == "celsius":
        return celsius
    elif scale == "kelvin":
        return celsius_to_kelvin(celsius)
    else:
        raise ValueError("scale must be 'celsius' or 'kelvin'")

print(from_fahrenheit(32))                     # -> 0.0 C
print(from_fahrenheit(212, scale="kelvin"))    # -> 373.15 K
````

### 5) Write a function that takes two arguments (feet and inches) and returns height in meters

Verify it gives sensible answers

````python
def height_to_meters(feet, inches):
    """Convert a height given in feet and inches to meters."""
    total_inches = feet * 12 + inches
    return total_inches * 0.0254

print(height_to_meters(5, 11))   # ~1.80 m
print(height_to_meters(6, 0))    # ~1.83 m
````

### 6) Write a function that takes one argument (height in meters) and returns two values (feet and inches)

(Consult the [tutorial on numbers](https://docs.python.org/3/tutorial/introduction.html#numbers) if you're stuck on how to implement this.)

````python
def meters_to_height(meters):
    """Convert a height in meters to (feet, inches)."""
    total_inches = meters / 0.0254
    feet = int(total_inches // 12)
    inches = total_inches - feet * 12
    return feet, inches

print(meters_to_height(1.80))
````

### 7) Verify that the "round trip" conversion from and back to meters is consistent 

Check for 3 different values of height in meters

````python
for m in [1.50, 1.75, 2.00]:
    feet, inches = meters_to_height(m)
    back = height_to_meters(feet, inches)
    print(f"{m} m -> {feet} ft {inches:.2f} in -> {back:.4f} m")
````
