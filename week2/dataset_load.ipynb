```python
import pandas as pd

url = "https://gml.noaa.gov/webdata/ccgg/trends/co2/co2_mm_mlo.csv"
cols = ["year", "month", "decimal_date", "average", "deseasonalized", "ndays", "sdev", "unc"]

df = pd.read_csv(url, comment="#", header=None, names=cols)
df = df[pd.to_numeric(df["year"], errors="coerce").notna()].reset_index(drop=True)
df = df.apply(pd.to_numeric, errors="coerce")
df = df.astype({"year": int, "month": int})
df.head()
```
