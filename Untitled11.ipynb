```python
# Part 1: Basic square function
def square(x):
    return x ** 2

print("Square of 4:", square(4))
print("Square of 5:", square(5))

# Part 2: Extended power function with a default argument
def power(x, n=2):
    return x ** n

# Verification as requested
print("power(3) gives:", power(3))
print("power(3, n=3) gives:", power(3, n=3))
```

```python
def convert_temp(celsius, kelvin=False):
    """Converts Celsius to either Fahrenheit or Kelvin."""
    if kelvin:
        return celsius + 273.15
    else:
        return (celsius * 9 / 5) + 32

# Testing the function with different inputs
print("20°C in Fahrenheit:", convert_temp(20))
print("0°C in Fahrenheit:", convert_temp(0))
print("20°C in Kelvin:", convert_temp(20, kelvin=True))
print("-40°C in Kelvin:", convert_temp(-40, kelvin=True))
```

```python
# Defining the class
class WeatherStation:
    def __init__(self, station_id, latitude, longitude):
        self.station_id = station_id
        self.latitude = latitude
        self.longitude = longitude

# Creating an instance of the class
my_station = WeatherStation("NYC-01", 40.71, -74.00)

# Accessing the attributes
print("Station ID:", my_station.station_id)
print("Latitude:", my_station.latitude)
print("Longitude:", my_station.longitude)
```
